```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
import re
import os
```

```python
housing_df = pd.read_csv('housing_case.csv')
```

```python
housing_df.head()
```

```python
case_df = housing_df[['address','price','info','city','state']]
case_df.head()
```

```python
case_df["int_price"]=''
case_df.head()
```

```python
no_nullhousing_df = case_df.dropna(how='any')
no_nullhousing_df.count()
```

```python
no_nullhousing_df["city"].value_counts()
```

```python
city_list = list(set(no_nullhousing_df['city']))
```

```python
for index in no_nullhousing_df.index:
    no_nullhousing_df.iloc[index]['int_price'] = int(no_nullhousing_df.iloc[index]['price'].replace(',',''))
```

```python
no_nullhousing_df.head()
```

```python
item_df = pd.DataFrame(columns = ["city", "state", 'max_price', 'min_price', 'mean_price'])
```

```python
for item in city_list:    
    temp_df = no_nullhousing_df[(no_nullhousing_df.city==item)]
    maxprice = temp_df['int_price'].max()
    minprice = temp_df['int_price'].min()
    meanprice = temp_df['int_price'].mean()
#     meanprice = 0
#     thestate = 0
    thestate = temp_df['state'].iloc[0]
    item_df = item_df.append(pd.DataFrame({'city': [item], 'state': [thestate], 'max_price': [maxprice], 'min_price': [minprice], 'mean_price': [meanprice]}))

    
```

```python
# reset index
item_df.reset_index(drop=True, inplace=True)

# Write location field
item_df["location"] = item_df["city"] + ", " + item_df["state"]

# rearrange columns
city_house = item_df[['location', 'max_price', 'min_price', 'mean_price']]

# convert datatypes
city_house['max_price'] = city_house['max_price'].astype(str).astype(int)
city_house['min_price'] = city_house['min_price'].astype(str).astype(int)

#pd.options.display.float_format = '{:,.0f}'.format
city_house.mean_price = city_house.mean_price.astype(int)
```

```python
#item_df.to_excel('study.xlsx', sheet_name= 'Sheet1')


outname = 'zillowsummary.csv'

outdir = './output_data'
if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)    

city_house.to_csv(fullname)
```
